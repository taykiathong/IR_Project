```python
# !pip3 install --user -U nltk
!python3 -m spacy download en_core_web_lg 
!python -m spacy link en_core_web_lg en


import numpy as np
import time
import truecase
import spacy
from sklearn.feature_extraction.text import CountVectorizer
from scipy.spatial.distance import cosine
import numpy as np
import string
import nltk
nltk.download('punkt')


#use spacy larger model, more accurate
nlp = spacy.load("en")

#QUERY
query_input = input("Enter query: ").lower()

#start time
start_time = time.time() 


query_no_punctuation = str(query_input).translate(str.maketrans('', '', string.punctuation))

#get truecase of word in query , i.e Uniqlo is uppercase
query_no_punctuation = truecase.get_true_case(query_no_punctuation)

##parse and tag the query using spacy pipeline
query_no_punctuation= nlp(query_no_punctuation)
print("query_no_punctuation", query_no_punctuation)
query_no_punctuation = query_no_punctuation.text.lower()
print("query_no_punctuation", query_no_punctuation)

# query_no_punctuation = [query_no_punctuation.text]
# query_no_punctuation = query_no_punctuation.lower()
# print("query_no_punctuation", query_no_punctuation)


#create alias 
df = df.dropna(axis=0, subset=['keywords_combined'])
pos_keywords_df = df['keywords_combined']


corpus = []
#init word weights dictionary
word_weights = {}

#per row in the column
for i in pos_keywords_df:
    #true case of the word 
    i = truecase.get_true_case(i)
    #parse and tag the each document in corpus using spacy pipeline
    i = nlp(i)
    
    
    #for word in row
    for token in i:
        #if the truecase of the word is a proper noun (for location and company) or noun(position)
        if token.pos_ == 'PROPN' or token.pos_ == 'NOUN':
            #convert to text then lower case to compare with other columns 
            token = token.text
            token = token.lower()
            
            #if the word is in the list containing organisations, locations and companies , increase its weights
            #important terms
            if token in res:
#                 print(token)
                #increase weightage of word
                word_weights[token] = 2

    #convert i to text and append it to corpus list
    i=i.text
    corpus.append(i)
#     print(len(corpus))

#lower case everything in word_weights dictionray to standardise
word_weights_lower = dict((k.lower(), v) for k, v in word_weights.items()) 
```

```python
def tf_(doc):
    frequencies = {} 
    for term in doc:
        term_count = frequencies.get(term, 0) + 1
        frequencies[term] = term_count 
    
#     print("frequencies1", frequencies)
    for key, value in frequencies.items():
        for k, v in word_weights_lower.items():
#             print("k", k)
            if key==k:
                frequencies[key] *= 2
            else:
                value == value
#     print("frequencies2", frequencies)
    
    return frequencies
```

```python
def df_(docs):
    df = {}
    for doc in docs:
        for term in set(doc):
            df_count = df.get(term, 0) + 1 
            df[term] = df_count
    return df
```

```python
def idf_(df, corpus_size):
    import math
    idf = {}
    for term, freq in df.items():
        idf[term] = round(math.log((corpus_size) / (freq)),2) 
    return idf

# print(idf_(df_(df['remove_stopwords']), len(df['remove_stopwords'])))
```

```python
def _score(query, doc, docs, k1=1.5, b=0.75): 
    score = 0.0
    tf = tf_(doc)
    df = df_(docs)
    idf = idf_(df, len(docs))
#     print("idf", idf)
    avg_doc_len = sum(len(doc) for doc in docs)/len(docs) 
    for term in query:
        if term not in tf.keys(): 
            continue
#         print(term)
        numerator = idf[term] * tf[term] * (k1 + 1)
#         print("numerator", numerator)
        denominator = tf[term] + k1 * (1 - b + b * len(doc) / avg_doc_len) 
#         print("denominator", denominator)
        score += (numerator / denominator)
    #
#     print(score)
    return score
```

```python
score = []
docs = df['lst_keywords_combined']
for i in range(len(docs)):
    doc = docs[i]
#     print(doc)
    bm25 = _score(query_no_punctuation, doc, docs, k1=1.5, b=0.75)
#     print(bm25)
#     bm25 = _score(query_input, doc, docs, k1=1.5, b=0.75)
    score.append(bm25)
# print(score)


end_time = time.time() - start_time

sorted_score_lst = sorted(score, reverse=True)
sorted_index = [score.index(i) for i in sorted_score_lst]
combined_dictionary = dict(zip(sorted_index, sorted_score_lst))



#carry out ranking of the documents
rank = 0 

for k,v in combined_dictionary.items():
    rank = rank + 1
    print("\n")
    print(f"Rank Number --> {rank}")
    print(f"Document is --> {corpus[k]}")
    print(f"Bm25 score is --> {v}")
    print("\n")
    
    #top ten
    if rank == 10:
        break


end_time = time.time() - start_time
print(f'Total Time taken = {end_time} seconds')
```

```python

top_results = pd.DataFrame(combined_dictionary.items(), columns=['Index', 'bm25_results'])

result_company = []
result_occupation = []
result_place = []
result_keywords = []

for i in range(len(top_results['Index'])):
    result_company.append(df['company_lower'][top_results['Index'][i]])
    result_occupation.append(df['occupation_lower'][top_results['Index'][i]])
    result_place.append(df['place_lower'][top_results['Index'][i]])
    result_keywords.append(df['keywords_combined'][top_results['Index'][i]])
    
top_results['Company'] = result_company
top_results['Occupation'] = result_occupation
top_results['Place'] = result_place
top_results['Keywords'] = result_keywords

top_results.head(10)
```
