```python
import pandas as pd
import nltk
nltk.download('stopwords')
nltk.download('punkt')
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import string

df = pd.read_csv("job_reviews.csv")
```

```python
df.index = [x for x in range(1, len(df.values)+1)]
df.index.name = 'id'
df.head(15)
```

```python
# review_lower = []
review_split_words = []

for index, row in df.iterrows():
    s = row['Review']
    s = s.lower()
#     review_lower.append(s)
    s = s.translate(str.maketrans('', '', string.punctuation))
    split_words = word_tokenize(s)
    review_split_words.append(split_words)
    
# df["review_lower"] = review_lower
df["review_split_words"] = review_split_words
```

```python
remove_stopwords = []
stop_words = set(stopwords.words('english'))

for index, row in df.iterrows():
    word_tokens = row['review_split_words']
    filtered_sentence = [w for w in word_tokens if not w.lower() in stop_words]
    remove_stopwords.append(filtered_sentence)
print(remove_stopwords[0])
    
    
df["remove_stopwords"] = remove_stopwords
df.head(10)
```

```python
string_remove_stopwords = []

for index, row in df.iterrows():
    words = row['remove_stopwords']
    sentence = ' '.join(words)
    sentence = sentence.lower()

    string_remove_stopwords.append(sentence)
    
    
df['string_remove_stopwords'] = string_remove_stopwords
df.head(10)
```

```python
companyLower = []

for index, row in df.iterrows():
    name = row['Company']
    name = name.lower()
    newname = name.replace('-', ' ')
    companyLower.append(newname)
    
    
df['company_lower'] = companyLower
df.head(10)
```

```python
occupationLower = df['Occupation'].map(lambda x: x.lower() if isinstance(x,str) else x)
df['occupation_lower'] = occupationLower

placeLower = df['Place'].map(lambda x: x.lower() if isinstance(x,str) else x)
df['place_lower'] = placeLower
    
df.head(10)
```

```python
df['keywords_combined'] = df['string_remove_stopwords'] + ' ' + df['occupation_lower'] + ' ' + df['company_lower'] + ' ' + df['place_lower']
df.head(10)
```

```python
lst_keywords_combined = []
for index, row in df.iterrows():
    s = str(row['keywords_combined'])
    s = s.split()
    lst_keywords_combined.append(s)

df["lst_keywords_combined"] = lst_keywords_combined
df.head(10)
```

```python
df = df.drop(columns = ['_OccupationRole','_Rank', 'Author', 'Company', 'Date', 'Employment', 'Occupation', 'Place','Review', 'Rating', 'review_split_words', 'remove_stopwords', 'lst_keywords_combined'])
df.head(10)
```

```python
df.to_csv("cleaned_data.csv")
```

```python
from sklearn.model_selection import train_test_split

train, test = train_test_split(df, test_size=0.2)


print(len(train))
print(len(test))
print(len(train)/len(df))
```
